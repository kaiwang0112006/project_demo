```python
def strLengthGreaterThen3(str):
    if len(str) > 3:
        tag = 1
    else:
        tag = 0
    return(tag)
```

```python
def isMobile(str):
    if len(str) > 11:
        tmp = str[-11:-1]
        if tmp[0] == '1':
            tag = 1
        else:
            tag = 0
    else:
        tag = 0
    return(tag)
```

```python
def getcommonSDK(df, ):
    '''
    输入一个dataframe，提取每个用户pby('在统计时间内使用的设备型号数，设备数量，设备的imei数量，设备的ip数量，设备的imis数量等
    '''
    df = df.drop_duplicates()
    contacts_size = df.groupby('ugid').phone.count().reset_index()        #对每个样本的通讯录长度进行去重计数
    contacts_size.columns = ['ugid', 'contacts_size']
    
    df['name'] = df['name'].astype('str')
    df['phonename_g3'] = df['name'].apply(lambda x:strLengthGreaterThen3(x))
    phonename_g3 = df.groupby('ugid').phonename_g3.count().reset_index()    #计算样本中存储的号码，有多少存储姓名的字符长度大于3位
    phonename = df.groupby('ugid').name.count().reset_index()
    
    df['phone'] = df['phone'].astype('str')
    df['mobile_contacts'] = df['phone'].apply(lambda x:isMobile(x))
    mobile_contacts = df.groupby('ugid').mobile_contacts.count().reset_index()
    phone_contacts = df.groupby('ugid').phone.count().reset_index()
    
    
       
    feature01 = df['ugid'].drop_duplicates().reset_index()
    feature01 = pd.merge(feature01, contacts_size, how = 'left', on = 'ugid')
    feature01 = pd.merge(feature01, phonename_g3, how = 'left', on = 'ugid')
    feature01 = pd.merge(feature01, phonename, how = 'left', on = 'ugid')
    feature01 = pd.merge(feature01, mobile_contacts, how = 'left', on = 'ugid')
    feature01 = pd.merge(feature01, phone_contacts, how = 'left', on = 'ugid')
    
    feature01['phonename_g3_rate'] = feature01['phonename_g3'] / feature01['name']
    feature01['mobile_contacts_rate'] = feature01['mobile_contacts'] / feature01['phone']
    
    feature01 = feature01[['ugid','contacts_size', 'phonename_g3_rate', 'mobile_contacts_rate']]
    
    return(feature01)
```

```python
feature01 = getcommonSDK(con01)
```
