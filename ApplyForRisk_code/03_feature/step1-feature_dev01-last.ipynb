```python
def getcommonSDK(df, days):
    '''
    输入一个dataframe，提取每个用户在统计时间内使用的设备型号数，设备数量，设备的imei数量，设备的ip数量，设备的imis数量等
    '''
    df = dev01[(dev01.ctime + timedelta(days=30) > dev01.last_decision_tm) &  (dev01.ctime<dev01.last_decision_tm) ].reset_index(drop=True)
    feature01 = df[['ugid','last_decision_tm']].drop_duplicates()
    
    tmp = df[['ugid', 'model']].drop_duplicates()
    model_cnt = tmp.groupby('ugid').model.count().reset_index()        #统计时间内设备型号去重个数
    tmp = df[['ugid', 'zuid']].drop_duplicates()
    device_cnt = tmp.groupby('ugid').zuid.count().reset_index()          #统计时间内设备去重个数
    tmp = df[['ugid', 'imei']].drop_duplicates()
    imei_cnt = tmp.groupby('ugid').imei.count().reset_index()          #统计时间内设备imei去重个数
    tmp = df[['ugid', 'remote_addr']].drop_duplicates()
    ip_cnt = tmp.groupby('ugid').remote_addr.count().reset_index()      #统计时间内设备ip去重个数
    tmp = df[['ugid', 'imsi']].drop_duplicates()
    imsi_cnt = df.groupby('ugid').imsi.count().reset_index()          #统计时间内设备imsi去重个数
    
    
    model_cnt.columns = ['ugid', 'last1m_model_cnt']
    device_cnt.columns = ['ugid', 'last1m_device_cnt']
    imei_cnt.columns = ['ugid', 'last1m_imei_cnt']
    ip_cnt.columns = ['ugid', 'last1m_ip_cnt']
    imsi_cnt.columns = ['ugid', 'last1m_imsi_cnt']
       
    feature01 = pd.merge(feature01, model_cnt, how = 'left', on = 'ugid')
    feature01 = pd.merge(feature01, device_cnt, how = 'left', on = 'ugid')
    feature01 = pd.merge(feature01, imei_cnt, how = 'left', on = 'ugid')
    feature01 = pd.merge(feature01, ip_cnt, how = 'left', on = 'ugid')
    feature01 = pd.merge(feature01, imsi_cnt, how = 'left', on = 'ugid')
    return(feature01)
```

```python
def recently_used_duration(x,df,days):
    tmp = df[df['ugid'] == x]
    tmp = tmp.sort_index(axis=0, by='ctime', ascending=False).reset_index()
    ctime = tmp['ctime'][0]
    #最近一次使用距今的时间
    last_time = tmp['last_decision_tm'][0]
    ctime = last_time - ctime
    return(ctime)
```

```python
def recentlyUsedfeature(feature, dataf, days):
    '''
    输入一个dataframe，提取每个用户在统计时间内最近使用的一个设备的统计特征，包括最近一个使用设备的使用时间，最近一次使用的设备类型，最近一次使用的设备os型号
    '''
    
    df = dataf[(dataf.ctime + timedelta(days) > dataf.last_decision_tm) & (dataf.ctime < dataf.last_decision_tm)].reset_index(drop=True)
    
    # 先排序
    df.sort_values(by=['ugid','ctime'],ascending=[True,False],inplace=True)
    
    # 再去重
    df=df.drop_duplicates(['ugid'])
    
    recent_used_device = df[['ugid','device_type','ctime','sysVersion']]
    recent_used_device.columns = ['ugid','recent_device_type','recent_device_change_days','recent_device_os_type_2']
    
    
    
    feature = pd.merge(feature, recent_used_device, how = 'left', on = 'ugid')
    
    return feature
```
