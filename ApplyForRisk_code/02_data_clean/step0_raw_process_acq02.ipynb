```python
'''
    此程序作用为从把筛选后的数据处理为可以直接输入变量衍生代码的csv格式
'''
import pandas as pd
import json
import numpy as np
```

```python
# 定义acq01-03,idcard原始文件列名
colums_name={
    'ACQ01': ['ugid', 'zuid', 'appid', 'utime', 'ctime', 'path', 'remote_addr', 'sessionId' ],
    'ACQ02': ['ugid' ,'zuid' ,'appid' ,'utime' ,'ctime' ,'ch' ,'eleId' ,'path' ,'remote_addr' ,'sessionId' ,'startTime' ,'endTime' ,'stayTime' ,'val'],
    'ACQ03': ['ugid' ,'zuid' ,'appid' ,'utime' ,'ctime' ,'ch' ,'clickTime' ,'eleId' ,'path' ,'remote_addr' ,'sessionId'],
    'IDCARD01': ['ugid' ,'zuid' ,'appid' ,'utime' ,'ctime' ,'address' ,'allTime' ,'birthday' ,'edit_address_time' ,'edit_birthday_time' 
             ,'edit_id_card_number_time' ,'edit_issuing_authority_time' ,'edit_name_time' ,'edit_period_of_validity_time' ,'edit_race_time' 
             ,'gender' ,'id_card_number_masked' ,'issued_by' ,'name_masked' ,'race' ,'remote_addr' ,'statistic_back_idcard_recognition_count' 
             ,'statistic_face_recognition_count' ,'statistic_front_idcard_recognition_count' ,'valid_date']
}
```

```python
data = []
#with open('../data_mid/data.201608_201702.screen_all.log', 'r') as f:
with open('../data_mid/20171122_data.201706_201708.screen_all.log', 'r') as f:
    for line in f:
        if line[0:4] != 'ugid':
            data.append(line.split('\t'))
```

```python
processed_data = []
a = 0
for item in data:
    try:
        ugid, zuid, appid, mid, utime = item[0:-1]
    except:
        print('item error:')
        print(item)
        continue
    if mid!='ACQ02':
        a+=1
    else:
#         print(item[-1].strip("'\r\n\""))
        body = json.loads(item[-1].strip("'\r\n\""))
        for sub_body in body:
            ctime ,ch ,eleId ,path ,remote_addr ,sessionId ,startTime ,endTime ,stayTime ,val = sub_body
            try:
                sessionId = str(int(ctime*1000))
            except:
                print('sub_body error:')
                print(sub_body)
                sessionId = np.nan
            processed_data.append([ugid, zuid, appid, utime, ctime ,ch ,eleId ,path ,remote_addr ,sessionId ,startTime ,endTime ,stayTime ,val])
```

```python
tmp_pd = pd.DataFrame(processed_data)
tmp_pd.columns = ['ugid', 'zuid', 'appid', 'utime', 'ctime','ch', 'eleId', 'path', 'remote_addr', 'sessionId', 'startTime', 'endTime', 'stayTime', 'val']
```
