```python
data = []
with open('../data_mid/20171122_data.201706_201708.screen_all.log', 'r') as f:
    for line in f:
        if line[0:4] != 'ugid':
            data.append(line.split('\t'))
```

```python
processed_data = []
a = 0
for item in data:
    try:
        ugid, zuid, appid, mid, unknwon = item[0:-1]
    except:
        print(item)
        continue
    if mid != 'IDCARD01':
        a+=1
    else:
        body = json.loads(item[-1].strip("'\r\n\""))
        for sub_body in body:
            ctime,address,allTime,birthday,edit_address_time,edit_birthday_time,edit_id_card_number_time,edit_issuing_authority_time,edit_name_time,edit_period_of_validity_time,edit_race_time,gender,id_card_number_masked,issued_by,name_masked,race,remote_addr,statistic_back_idcard_recognition_count,statistic_face_recognition_count,statistic_front_idcard_recognition_count,valid_date = sub_body
            sessionId = str(int(ctime*1000))
            utime = ctime
            processed_data.append([ugid,zuid,appid,utime,ctime,address,allTime,birthday,edit_address_time,edit_birthday_time,edit_id_card_number_time,edit_issuing_authority_time,edit_name_time,edit_period_of_validity_time,edit_race_time,gender,id_card_number_masked,issued_by,name_masked,race,remote_addr,statistic_back_idcard_recognition_count,statistic_face_recognition_count,statistic_front_idcard_recognition_count,valid_date])
```
